## 機能要件

- デプロイチェックリストクリア

- 管理人に連絡をとれる。コンタクトフォーム。通報窓口

- 管理者アカウント
  - BAN権限


- ユーザアカウント
  - ニックネーム、メールアドレス、投稿コンテンツ、を紐づける
  - ユーザは自分のコンテンツを削除できる
  - 投稿ユーザごとの一覧表示
  - アカウント削除すると投稿コンテンツも削除される


- コンテンツ
  - 写り込みフィルタ、人の写り込みは投稿不可
  - 撮影場所
  - 日時
  - カテゴリ、タグ付け、むしろないほうがよいかも
  - ランダム再生、意外性
  - 再生履歴
  - 情報参照できる


- デバイス
  - PC, モバイルレスポンシブ
  - モバイルからも投稿可能
